```python
import lightkurve as lk
from astropy.timeseries import BoxLeastSquares
import numpy as np
import astropy.units as u
from astroquery.mast import Observations
from astropy.io import fits
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt
```

```python


epicID = "211359660"
with fits.open('mastDownload/HLSP/hlsp_everest_k2_llc_211359660-c05_kepler_v2.0_lc/hlsp_everest_k2_llc_211359660-c05_kepler_v2.0_lc.fits') as hdu:
    asn_table = Table(hdu[1].data)

#print(asn_table[0])
'''
FRAW = Raw version of plot
FLUX = The corrected version
'''
x = asn_table['CADN']
y = asn_table['FLUX']

# Convert y to np.arr
yArr = np.asarray(y)
xArr = np.asarray(x)
print(yArr)
```

```python

plt.figure(num=None, figsize=(16, 8), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(x,y,s = 1 )
```

```python
lc = lk.LightCurve(time=xArr, flux=yArr)
#lc = lc.flatten(window_length=1001).remove_outliers()
lc.plot();
```

```python
flat, trend = lc.flatten(window_length=301, return_trend=True)
```

```python
ax = lc.errorbar(label=epicID)                   # plot() returns a matplotlib axes ...
trend.plot(ax=ax, color='red', lw=2, label='Trend');  # which we can pass to the next plot() to use the same axes
```

```python
flat.errorbar(label=epicID);
```

```python
periodogram = flat.to_periodogram(method="bls", period=np.arange(0.5, 1.5, 0.001))
periodogram.plot();
```

```python
best_fit_period = periodogram.period_at_max_power
print('Best fit period: {:.3f}'.format(best_fit_period))
```

```python
flat.fold(period=best_fit_period, epoch_time=periodogram.transit_time_at_max_power).errorbar();
```
